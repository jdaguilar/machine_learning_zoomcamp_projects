```python
plt.figure()
plt.scatter(df["passenger_count"], df['fare_amount'], alpha=0.5)
plt.title(f'Scatter Plot of {"passenger_count"} vs fare_amount')
plt.xlabel("passenger_count")
plt.ylabel('fare_amount')
plt.grid()
plt.show()
```

```python
df = df[df["passenger_count"] != 208]
```

```python
sns.violinplot(data=df, x="passenger_count", y="fare_amount", fill=False)
```
